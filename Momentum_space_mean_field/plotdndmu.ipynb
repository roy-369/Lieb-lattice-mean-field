```python
# flat-band stuff
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
from matplotlib import cm

import seaborn as sns
import pandas as pd
sns.axes_style("whitegrid")

from tqdm import tqdm

import matplotlib
matplotlib.rcParams['text.usetex'] = True

font = {'family' : 'Times New Roman',
        'weight' : 'bold',
        'size'   : 12}

matplotlib.rc('font', **font)
```

```python
data = pd.read_csv('fb-nmud.csv', header=None)
    
data.columns = ["mu", "den"]

```

```python
muar = data.mu.to_numpy()
denar = data.den.to_numpy()
dndu = denar
```

```python
denar
```

```python
muar[55]
```

```python

for ctr in range(len(muar)):
    d1 = denar[ctr]
    m1 = muar[ctr]
    if (ctr < len(data.den.to_numpy())-1):
        d2 = denar[ctr+1]
        m2 = muar[ctr+1]
        
        nmr = d2-d1
        dnr = m2-m1
        dndu[ctr] = nmr/dnr
        print(d1,d2, dnr)
    else:
        dndu[ctr] = 0.0
        
dndu[0] = 0.0
    
```

```python
np.shape(dndu)
```

```python
fig, ax = plt.subplots(1,1,figsize=(5,4))
fig.tight_layout()
ax.set_xlabel(r'$\mu$')
ax.set_ylabel(r'$n$')
#ax.set_title(r'Square Lattice')

factr = 1.e3/np.pi

ax.grid(linestyle="--", linewidth=0.3, color=".25", zorder=-10)


#plt.plot(muar,denar,label= r'n',marker='s',markersize=3)
plt.plot(muar,dndu,label= r'$dn/d\mu$',marker='o',markersize=3)

plt.legend(fontsize=14,loc='lower right')
```
