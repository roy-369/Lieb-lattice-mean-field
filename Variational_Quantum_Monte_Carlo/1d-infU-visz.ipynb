```python
import numpy as np
import seaborn as sns
sns.axes_style("whitegrid")
import matplotlib.pyplot as plt
```

```python
dg = 0.1
Ns = 100
Np = 10

fname_suffix = ('_dg%3.2f_fl_%3.2f'%(dg, float(Np)/Ns)).replace('.','p')

PE_g = np.load('data/PEg'+fname_suffix+'.npy')
KE_g = np.load('data/KEg'+fname_suffix+'.npy')
KE_err = np.load('data/KEg_err'+fname_suffix+'.npy')
PE_err = np.load('data/PEg_err'+fname_suffix+'.npy')
gar = np.arange(0,1+dg,dg)
```

```python
plt.errorbar(gar, PE_g, yerr=PE_err )
plt.scatter(gar, PE_g)
plt.show()
```

```python
plt.errorbar(gar, KE_g, yerr=np.multiply(KE_err,1.0))
plt.scatter(gar, KE_g)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(3,9),nrows=3, sharex=True)

U = 5
TE_g = KE_g + U*PE_g
TE_err = (1.0/95)*(KE_err + U*PE_err)
ax[0].errorbar(gar, TE_g, yerr=TE_err)
ax[0].scatter(gar, TE_g)
ax[0].axhline(y=np.min(TE_g),linestyle='dashed')
ax[0].set_ylabel('E = T + V')
ax[0].set_title('U=5.0')

U = 0.1
TE_g = KE_g + U*PE_g
TE_err = (1.0/95)*(KE_err + U*PE_err)
ax[1].errorbar(gar, TE_g, yerr=TE_err)
ax[1].scatter(gar, TE_g)
ax[1].axhline(y=np.min(TE_g),linestyle='dashed')
ax[1].set_xlabel('g')
ax[1].set_ylabel('E = T + V')
ax[1].set_title('U=0.5')

U = 0.0
TE_g = KE_g + U*PE_g
TE_err = (1.0/95)*(KE_err + U*PE_err)
ax[2].errorbar(gar, TE_g, yerr=TE_err)
ax[2].scatter(gar, TE_g)
ax[2].axhline(y=np.min(TE_g),linestyle='dashed')
ax[2].set_xlabel('g')
ax[2].set_ylabel('E = T + V')
ax[2].set_title('U=0.0')

fig.tight_layout(pad=1.0)

plt.show()

#fig.savefig('figs/energy_vs_g_vqmc.jpg',bbox_inches='tight', dpi=300)
```

```python
fig, ax = plt.subplots()

U = 0.0
TE_g = KE_g + U*PE_g
TE_err = (1.0/95)*(KE_err + U*PE_err)
ax.errorbar(gar, TE_g, yerr=TE_err)
ax.scatter(gar, TE_g)
ax.axhline(y=np.min(TE_g),linestyle='dashed')
ax.set_xlabel('g')
ax.set_ylabel('K.E')

plt.show()

#fig.savefig('figs/energy_vs_g_vqmc.jpg',bbox_inches='tight', dpi=300)
```
